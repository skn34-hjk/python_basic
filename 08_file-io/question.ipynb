```python
import csv

data = [
    ["회사명","직무","결과"],
    ["회사A","데이터 분석가","합격"],
    ["회사B","소프트웨어 엔지니어","불합격"],
    ["회사C","제품 관리자","진행 중"]
]

with open('applications.csv', 'w', encoding = 'utf-8') as csvfile:
    w = csv.writer(csvfile,delimiter=",")
    for i in data:
        w.writerow(i)
    
        print(i)
```

```python
import json

data = {
    "name": "카피바라",
    "favorite_foods": ["도토리", "호두", "밤"]
}

with open('favorite_foods.json', 'w', encoding = 'utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=4) #json은 dump로 저장

with open('favorite_foods.json', 'r', encoding = 'utf-8') as f:
    json_dict = json.load(f) # json을 딕셔너리로 자동 형변환 시켜주는 json.load()
    print(json_dict["favorite_foods"])
```
