```python
import csv

data = [
    ["회사명","직무","결과"],
    ["회사A","데이터 분석가","합격"],
    ["회사B","소프트웨어 엔지니어","불합격"],
    ["회사C","제품 관리자","진행 중"]
]

with open('applications.csv', 'w', encoding = 'utf-8') as csvfile:
    w = csv.writer(csvfile,delimiter=",")
    for i in data:
        w.writerow(i)
    
        print(i)
```

```python
with open('system.log', 'w', encoding = 'utf-8') as f:
    log_content = [
        '[INFO] 2026-01-01 12:00:00: 시스템 시작\n',
        '[WARN] 2026-01-01 12:05:00: 메모리 사용량 증가\n',
        '[ERROR] 2026-01-01 12:10:00: 프로그램 종료'
    ]
    f.writelines(log_content)
```

```python
with open('system.log', 'r', encoding = 'utf-8') as f:
    for line in f:
        if 'WARN' in line or 'ERROR' in line:
            print(line)
```

```python
import json

data = {
    "name": "카피바라",
    "favorite_foods": ["도토리", "호두", "밤"]
}

with open('favorite_foods.json', 'w', encoding = 'utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=4) #json은 dump로 저장

with open('favorite_foods.json', 'r', encoding = 'utf-8') as f:
    json_dict = json.load(f)
    print(json_dict["favorite_foods"])
```

```python
with open('resume.html', 'w', encoding = 'utf-8') as f:
    f.write("""
<html>
    <head><title>카피바라의 이력서</title></head>
    <body>
        <h1>카피바라의 이력서</h1>
        <p>경력: 신입</p>
    </body>
</html>
    """)

f.close()
```

```python
with open('resume.html', 'r', encoding = 'utf-8') as f:
    content = f.read()
    
modifuded_content = content.replace('신입', '1년 경력')

with open('resume.html', 'w', encoding = 'utf-8') as f:
    f.write(modifuded_content)

with open('resume.html', 'r', encoding = 'utf-8') as f:
    print(f.read())
```
