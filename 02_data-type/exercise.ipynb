```python
from collections import deque

dq = deque()

dq.append("a")
dq.append("b")
dq.append("c")

print(dq)    # deque(['a', 'b', 'c']) 
```

```python
dq.appendleft('x')
print(dq)   # deque(['x', 'a', 'b', 'c'])
```

```python
value = dq.pop()
print(value)    # c

left_value = dq.popleft() # 왼쪽 끝 함수 반환하며 삭제
print(left_value)    # x

print(dq)    # deque(['a', 'b'])
```

```python
javaTeam = {'Squirrel','Gorilla', 'Tiger', 'Sheep', 'monkey', 'wolf'}
pythonTeam = {'Pig', 'Squirrel', 'Bear', 'Gorilla'}

#합집합

ohgiraffers = javaTeam.union(pythonTeam)
print(ohgiraffers)

print(javaTeam | pythonTeam)
```

```python
# 교집합

print(javaTeam.intersection(pythonTeam))

print(javaTeam & pythonTeam)
```

```python
#차집합

print(javaTeam.difference(pythonTeam))

print(javaTeam - pythonTeam)
```

```python
# 공통차집합, 대칭차집합 = 공통된 부분을 뺀 거 = 합집합 - 교집합

print(javaTeam ^ pythonTeam)                        # {'wolf', 'Sheep', 'Tiger', 'monkey', 'Pig', 'Bear'}
print(javaTeam.symmetric_difference(pythonTeam))    # {'wolf', 'Sheep', 'Tiger', 'monkey', 'Pig', 'Bear'}
```

```python
k = ['one', 'two', 'three']
v = [1,2,3]

# 두개의 자료를 순서대로 쌍으로 묶은 zip객체를 반환
t = zip(k,v)
print(t) #<zip object at 0x034ABE90>
print(type(t)) #<class 'zip'>

for k,v in t:
    print(k,v)
# one 1
# two 2
# three 3
```

```python
# 1.shallow copy = 얕복 = 메모리상의 주소가 복사됨. (같은 주소를 가짐.)
a = {'a': 1, 'b': 2, 'c': 3}
b = a # 사전 레퍼런스의 복사 (공유됨)
print(a)
print(b)

a['a'] = 100
print(a)
print(b)
```

```python
# 2.deep copy = 깊복 = 값을 새로 만듦. 주소도 다름.
c = {'a': 1, 'b': 2, 'c': 3}
d = c.copy()
print(c)
print(d)

c['a'] = 300
print(c)
print(d) # 변경되지 않음.
```

```python
import copy

origin = {
    'name' : 'capybara', 
    'score' : [100, 90, 80]
}

deep = copy.deepcopy(origin)

origin['score'][0] = 999

print(origin)
print(deep)
```

```python
# update 메소드를 통한 두 dict의 병합
# 동일한 키값이 있다면, 인수로 주어진 dict의 아이템으로 덮어쓰기 된다.
dic1 = {'a':10, 'b':20}
dic2 = {'c':100, 'd':300}
dic1.update(dic2)

print(dic1) #{'a': 10, 'b': 20, 'c': 100, 'd': 300}
print(dic2) #{'c': 100, 'd': 300}

# 딕셔너리 두 개를 합칠 때는, 업데이트를 사용.
```

```python
# dict의 정렬 : sorted
numbers = {'first': 1, 'second': 2, 'third': 3, 'Fourth': 4}
numbers['hundred'] = 100
numbers['fifth'] = 5

print(list(numbers)) #['first', 'second', 'third', 'Fourth', 'hundred', 'fifth']

# key값 정렬
print(sorted(numbers)) #['Fourth', 'fifth', 'first', 'hundred', 'second', 'third']
# sorted() 는 결과값이 리스트로 나온다. 

# value값 정렬
print(sorted(numbers.values())) #[1, 2, 3, 4, 5, 100]

# key정렬 (value값 정렬에 따라)
print(sorted(numbers, key=numbers.__getitem__)) # key 는 item 을 기준으로 정렬
```
