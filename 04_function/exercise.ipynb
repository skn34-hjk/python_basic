so, 키워드 인자로 작성 or 위차 맞추어 작성

```python
def introduce(name):
    return f'안녕하세요, 저는 {name}입니다-!'

result = introduce('hyeonn')
print(result)
```

```python
def calc(a, b):
    return a+b, a-b

print(calc(2, 3))
print(type(calc(2, 3))) # 패킹되어 튜플 형태로 출력
```

```python
plus, minus = calc(5, 7)

print(f'plus는 {plus}, minus는 {minus}.') # 언패킹
```

```python
# *는, 매개변수 인자값의 개수가 정해지지 않았을 때 사용.
# if 키워드 인자인 경우 *이거 두 개를 붙여서 작성 **

def add_many(*args):
    result = 0

    for i in args:
        result = result + i

    return result

print(add_many(1, 2, 3, 4, 5))
```

```python
def show(*args):
    print(args)

show(1, 2, 3, 'cap') # 튜플 형태로 출력
```

```python
def show_info(**kwargs):
    print(kwargs)

show_info(name= 'capybara', age = 60) # 딕셔너리 형태로 출력
```

```python
def sum(a, b, c):
    return a+b+c

numbers = [1, 2, 3]
# sum([numbers]) # 리스트 형태의 값이 하나만 있는 것으로 인식.

print(sum(*numbers))  # sum(1, 2, 3) 언패킹
```

변수명 = lambda(매개변수) : (함수 실행 내용)

```python
lambda x: x + 1
lambda x: x ** 2
lambda x: x[1]

lambda x: (x+1 if x>0 else x-1)
```

```python
add = lambda x, y: x+y
print(add(3, 5))
```

```python
def add(x, y):
    return x + y

print(add(3, 5))
```

```python
numbers = [1, 2, 3, 4, 5]
squared_numbers = map(lambda x: x**2, numbers)
print(list(squared_numbers))
```

```python
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = filter(lambda x: x%2==0, numbers)
print(list(even_numbers))
```

```python
points = [(1, 2), (3, 1), (5, -1)]
sorted_points = sorted(points, key=lambda x: x[1]) # 두 번째 요소를 기준으로 오름차순 정렬
print(sorted_points)
```

```python
students = [
    {"name":"bear", "score":90},
    {"name":"tiger", "score":70},
    {"name":"rabbit", "score":85}
]

# 점수 기준 정렬
print(sorted(students, key=lambda x: x["score"]))
```
