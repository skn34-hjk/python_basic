````python
# 기본 구조는 try-except
````

```
 try:   
     (실행할 코드)   
 except:   
      (모든 예외 발생 시 동작할 코드)
```

````python
try:
    4/0
except: 
    print("예외 발생")
````

````python
try:
    a = [1,2]
    print(a[3])
    4/0
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다")
except IndexError:
    print("인덱싱 할 수 없습니다.")
````

````python
try:
    a = [1,2]
    print(a[3])
    4/0
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다")
except IndexError:
    print("인덱싱 할 수 없습니다.")
except:
    print('아무튼 에러')
````

````python
try:
    4/0
except ZeroDivisionError as e:
    print(e)
````

````python
f = None

try:
    f = open('foo.txt', 'r')
    # 동작할 코드 내용 작성

finally: #오류가 발생하던안하던 실행됨
    if f:
        f.close()   # try 블럭 수행 중간에 오류가 발생하더라도 마지막에 무조건 실행된다.
        print('정상 실행 완료')
````

````python
try:
    age = int(input('나이를 입력하세요: '))
except:
    print('입력이 정확하지 않습니다.')
else:
    if age <= 18:
        print('미성년자')
    else:
        print('성인')
````
