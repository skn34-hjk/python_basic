```python
# 실습문제 : Life is too short, you need python에서 aeiou제거하기

# 방법1
s = 'Life is too short, you need python'
print([x for x in s if(x not in ['a','e','i','o','u'])])

# 방법2
vowels = 'aeiou'
print(''.join([a for a in s if a not in vowels]))
```

```python
for x in [1, 2, 3]:
    
    print("앞", x)
    
    if x == 2:
        print("if", x)
        print("끝")  	
        break       # 반복문 전체 종료 -> 현재 돌고있는 해당 for문을 종료시킴.
    
    print("뒤", x)
```

```python
for x in range(3):	
    print("앞", x)
    
    if x == 1:
        print("if", x)
        print("continue 등장!")   # 해당 회차 반복문은 그대로 끝낫다 치고 ..  다음 반복문으로 넘김.	
        continue
    
    print("뒤", x)
```
