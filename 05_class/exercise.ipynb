```python
g_variable = "global variable" # => 전역변수 = 프로그램 상에서 전체에서 통용되는 변수

def modify_global():
    global g_variable
    g_variable = "global modified"

print(g_variable)    # global variable
modify_global()
print(g_variable)    # global modified
```

```python
def outer_function():
    variable = "outer variable"

    def inner_function():
        nonlocal variable # 한단계 위 사용 원할 때 nonlocal 사용
        variable = "modified"

    print(variable)
    inner_function()
    print(variable)

outer_function()
```

```python
class Character:
    def __init__(self, nickname, type):
        self.nickname = nickname      # Public attribute
        self.__type = type            # Private attribute

    def get_type(self):
        return self.__type

    def set_type(self, type):
        character_types = ['전사', '법사', '치유사']
        if type in character_types:
            self.__type = type
        else:
            raise ValueError("잘못된 캐릭터 타입입니다.")

my_character = Character("카피바라GO", "전사")

print(my_character.nickname)    # 카피바라GO

# Private attribute에 직접 접근 시도 (실패)
try:
    print(my_character.__type)    # 'Character' object has no attribute '__type'
except AttributeError as e:
    print(e)

# Private attribute에 접근하는 public method 사용
print(my_character.get_type())    # 전사

# Public method를 통해 private attribute 수정
my_character.set_type("법사")
print(my_character.get_type())    # 법사

# 잘못된 값으로 수정 시도
try:
    my_character.set_type("흑마법사")    # 잘못된 캐릭터 타입을 선택하였습니다.
except ValueError as e:
    print(e)
```

```python
class Person:    
    national = 'korea'
    language = 'korean'

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greeting(self): # self는 객체 자기 자신을 가리킴
        return '안녕하세요'
    
    def information(self):
        return 'im from '+self.national+' and i use '+self.language
    
    def fav(self, color):
        return 'i love '+color

squirrel = Person('다람쥐', 900)
capybara = Person('카피바라', 20)

print(squirrel)    # <__main__.Person object at 0x000002E16EE94560> 메모리상의 주소값이 나온다.
print(capybara)     # <__main__.Person object at 0x000002E16EE94BF0>

print(squirrel.national)            # korea
print(squirrel.greeting())          # 안녕하세요
print(squirrel.information())        
print(capybara.information())        # I'm from korea and I use korean. My name is 카피바라. I'm 20
print(capybara.fav('black'))    # I love black
```

상속 ; 기존의 클래스를 재사용해 새로운 클래스를 만드는 것.

class 클래스명(상속받을클래스명)

```python
class Person:
    nationl = 'kor' # 클래스 속성

    def greeting(self): # 클래스 메서드 정의
        return 'hello im python'
    
class Student(Person):
    pass

student = Student()
print(student.nationl)
print(student.greeting())
```

```python
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, name, grade):
        super().__init__(name) # super()는 부모를 가리킨다. -> 상위클이 갖고 있던 name을 상속받아 쓰기 위해. 
        self.grade = grade

student = Student('바라', '2학년')
print(student.name)
print(student.grade)
```

```python
class Person:
    nat = 'kor'

    def greeting(self):
        return 'hello this is python'
    
class Learner:
    def greeting(self):
        return 'hello this is Learner'
    
    def learn(self):
        return 'im learnig python'
    
class Student(Person, Learner):
    pass

student = Student()
print(student.greeting())
print(student.learn())
```

```python
class Person:
    nat = 'kor'

    def greeting(self):
        return 'hello this is python'
    
class Learner(Person):
    def __init__(self, subject):
        self.subject = subject
    
    def learn(self):
        return 'im learnig ' + self.subject
    
class Student(Learner):
    def __init__(self, name, subject):
        super().__init__(subject)
        self.name = name

    def greeting(self): # 재정의 = 오버라이딩
        return 'Hello, my name is ' + self.name
    
student = Student('카피바라', 'python')
print(student.greeting())
print(student.learn())
```

```python
class Person:

    def greeting(self):
        return '안녕하세요'
    
class Student(Person):
    def greeting(self):
        return super().greeting() + '. 저는 학생입니다.'
    
student = Student()
print(student.greeting())
```

```python
class Animal:
    def do(self):
        print('👉👈')
class Dog(Animal):
    def do(self):
        print('🐶🐶🐶')
class Duck(Animal):
    def do(self):
        print('🦆🦆🦆')
class Fish(Animal):
    pass    # do 호출시 부모 클래스 Animal#do가 호출된다.

for each in (Dog(), Duck(), Fish()):
    each.do()
```

다형성 = 같은 클래스 이름을 상속해도, 재정의를 해서 다 다르게 사용.   객체의 종류에 따라 행동(동작)이 다 달라진다.

```python
# 상속 관계가 아니더라도 동일한 이름의 함수가 있는 경우 다형성처럼 사용이 가능
class Alarm:
    def do(self):
        print('⏰⏰⏰')

for each in (Dog(), Duck(), Fish(), Alarm()):
    each.do()
```
